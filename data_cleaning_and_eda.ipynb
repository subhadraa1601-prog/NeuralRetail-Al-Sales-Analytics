```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

# Load both sheets
df_2009 = pd.read_excel(
    '../data/raw/online_retail_II.xlsx',
    sheet_name='Year 2009-2010'
)

df_2010 = pd.read_excel(
    '../data/raw/online_retail_II.xlsx',
    sheet_name='Year 2010-2011'
)

# Combine datasets
df = pd.concat([df_2009, df_2010], ignore_index=True)

# Check shape
print(df.shape)

# Preview
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

## Remove Missing Customer IDs


```python
df = df.dropna(subset=['Customer ID'])
```

## Remove Missing Descriptions


```python
df = df.dropna(subset=['Description'])
```

## Remove Negative Quantities


```python
df = df[df['Quantity'] > 0]
```

## Remove Invalid Prices


```python
df = df[df['Price'] > 0]
```

```python
df.head()
```

```python
print(df.shape)
```

```python
df = df.drop_duplicates()
```

```python
print(df.shape)
```

```python
df['Customer ID'] = df['Customer ID'].astype(int)
```

```python
df['TotalAmount'] = df['Quantity'] * df['Price']
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['MonthName'] = df['InvoiceDate'].dt.month_name()
df['Quarter'] = df['InvoiceDate'].dt.quarter
df['Weekday'] = df['InvoiceDate'].dt.day_name()
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
df.head()
```

```python
print(df.shape)
```

```python
df.to_csv('../data/processed/cleaned_retail_data.csv', index=False)
```

```python
total_revenue = df['TotalAmount'].sum()
total_orders = df['Invoice'].nunique()
total_customers = df['Customer ID'].nunique()
total_products = df['StockCode'].nunique()

print("Total Revenue:", total_revenue)
print("Total Orders:", total_orders)
print("Total Customers:", total_customers)
print("Total Products:", total_products)
```

```python
monthly_sales = df.groupby('MonthName')['TotalAmount'].sum().sort_values()

monthly_sales.plot(kind='bar', figsize=(12, 6))
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
top_countries = df.groupby('Country')['TotalAmount'].sum(
).sort_values(ascending=False).head(10)

top_countries.plot(kind='bar', figsize=(12, 6))
plt.title('Top 10 Countries by Revenue')
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = df.groupby('Description')['Quantity'].sum(
).sort_values(ascending=False).head(10)

top_products.plot(kind='barh', figsize=(12, 6))
plt.title('Top Selling Products')
plt.xlabel('Quantity Sold')
plt.show()
```

```python
weekday_sales = df.groupby('Weekday')['TotalAmount'].sum()

weekday_sales.plot(kind='bar', figsize=(10, 5))
plt.title('Sales by Weekday')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
hourly_sales = df.groupby('Hour')['TotalAmount'].sum()

hourly_sales.plot(figsize=(12, 6))
plt.title('Hourly Sales Trend')
plt.xlabel('Hour')
plt.ylabel('Revenue')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalAmount'], bins=50)
plt.title('Revenue Distribution')
plt.show()
```

```python
customer_spending = df.groupby('Customer ID')['TotalAmount'].sum()

plt.figure(figsize=(10, 6))
sns.histplot(customer_spending, bins=50)
plt.title('Customer Spending Distribution')
plt.show()
```
