# Power BI KPI Tables Preparation


```python
import pandas as pd

df = pd.read_csv('../data/processed/cleaned_retail_data.csv')
```

## Monthly Sales Summary


```python
monthly_summary = df.groupby(
    ['Year', 'Month', 'MonthName']
).agg({
    'TotalAmount': 'sum',
    'Invoice': 'nunique',
    'Customer ID': 'nunique',
    'Quantity': 'sum'
}).reset_index()

monthly_summary.columns = [
    'Year',
    'Month',
    'MonthName',
    'Revenue',
    'TotalOrders',
    'UniqueCustomers',
    'TotalQuantity'
]

monthly_summary.head()
```

```python
monthly_summary.to_csv(
    '../data/processed/monthly_summary.csv',
    index=False
)
```

## Country Sales Summary


```python
country_summary = df.groupby('Country').agg({
    'TotalAmount': 'sum',
    'Invoice': 'nunique',
    'Customer ID': 'nunique'
}).reset_index()

country_summary.columns = [
    'Country',
    'Revenue',
    'Orders',
    'Customers'
]

country_summary = country_summary.sort_values(
    by='Revenue',
    ascending=False
)

country_summary.head()
```

```python
country_summary.to_csv(
    '../data/processed/country_summary.csv',
    index=False
)
```

## Product Performance Summary


```python
product_summary = df.groupby(
    ['StockCode', 'Description']
).agg({
    'Quantity': 'sum',
    'TotalAmount': 'sum',
    'Invoice': 'nunique'
}).reset_index()

product_summary.columns = [
    'StockCode',
    'Description',
    'TotalQuantitySold',
    'Revenue',
    'OrderCount'
]

product_summary = product_summary.sort_values(
    by='Revenue',
    ascending=False
)

product_summary.head()
```

```python
product_summary.to_csv(
    '../data/processed/product_summary.csv',
    index=False
)
```

## Customer Summary


```python
customer_summary = df.groupby('Customer ID').agg({
    'TotalAmount': 'sum',
    'Invoice': 'nunique',
    'Quantity': 'sum',
    'InvoiceDate': ['min', 'max']
})

customer_summary.columns = [
    'TotalSpent',
    'TotalOrders',
    'TotalQuantity',
    'FirstPurchase',
    'LastPurchase'
]

customer_summary = customer_summary.reset_index()

customer_summary.head()
```

```python
customer_summary['FirstPurchase'] = pd.to_datetime(
    customer_summary['FirstPurchase']
)

customer_summary['LastPurchase'] = pd.to_datetime(
    customer_summary['LastPurchase']
)
```

```python
customer_summary['CustomerLifetimeDays'] = (
    customer_summary['LastPurchase']
    - customer_summary['FirstPurchase']
).dt.days
```

```python
customer_summary.to_csv(
    '../data/processed/customer_summary.csv',
    index=False
)
```

## Daily Sales


```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

daily_sales = df.groupby(
    df['InvoiceDate'].dt.date
).agg({
    'TotalAmount': 'sum'
}).reset_index()

daily_sales.columns = ['Date', 'Revenue']

daily_sales.head()
```

```python
daily_sales.to_csv(
    '../data/processed/daily_sales.csv',
    index=False
)
```
